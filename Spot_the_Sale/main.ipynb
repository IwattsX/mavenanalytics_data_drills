```python
import pandas as pd
```

```python
orders_df = pd.read_csv("orders.csv", parse_dates=["order_date"])
promotions_df = pd.read_csv("promotions.csv", parse_dates=["start_date", "end_date"])

print(orders_df.head())
print(promotions_df.head())
```

```python
new_df = pd.merge_asof(
    orders_df.sort_values(by='order_date'),
    promotions_df.sort_values(by="start_date"),
    left_on="order_date",
    right_on="start_date",
    direction="backward"
).query("order_date <= end_date")
new_df
```
